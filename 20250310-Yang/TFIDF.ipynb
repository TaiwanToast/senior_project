```python
import pypdf
import os
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
```

```python
pdf_content = []

def get_pdf_text(pdf_path):
    for files in os.listdir(pdf_path):
        if files.endswith('.pdf'):
            pdf = pypdf.PdfReader(open(pdf_path + '/' + files, 'rb'))
            text = ''
            for page in pdf.pages:
                # text += page.extract_text()
                pdf_content.append({
                    'file': files,
                    'text': page.extract_text()
                })
```

```python
# 對文本分詞
from ckip_transformers.nlp import CkipWordSegmenter

ws_driver = CkipWordSegmenter(model='bert-base', device=0)
```

```python
get_pdf_text('..\\product infomation')
query = '我最近想看推理小說'
ws_query = ws_driver([query])
# ws2 = [pdf_content[i].get('text') for i in range(len(pdf_content))]
print(ws_query)
# print(ws2)
```

```python
ws_pdf = ws_driver([pdf_content[i].get('text') for i in range(len(pdf_content))], use_delim=True)
print(ws_pdf)
```

```python
new_ws_pdf = []
for i in range(len(ws_pdf)):
    new_ws_pdf.append(' '.join(ws_pdf[i]))

for i in range(len(new_ws_pdf)):
    for j in range(5):
        print(new_ws_pdf[i][j], end='')
    print()
# len(new_ws_pdf)
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(new_ws_pdf)
```

```python
# 取得特徵名稱（詞彙表）
feature_names = vectorizer.get_feature_names_out()

# 轉為密集矩陣
dense_matrix = tfidf_matrix.toarray()
```

```python
# 顯示 TF-IDF 結果
for doc_idx, doc in enumerate(dense_matrix):
    print(f"\n文本 {doc_idx + 1}:")
    for word_idx, score in enumerate(doc):
        if score > 0:
            print(f"  {feature_names[word_idx]}: {score:.4f}")
```

```python
# 查看某個詞在第 1 篇文件中的 TF-IDF 分數
word = "書桌"
index = feature_names.tolist().index(word)
print(f"TF-IDF for '{word}':", dense_matrix[0][index])
```

```python
target_word = "耳麥"
# 確認詞是否存在
if target_word in feature_names:
    word_index = feature_names.tolist().index(target_word)
    
    print(f"詞語：'{target_word}' 在各文件的 TF-IDF 分數:")
    for doc_idx, doc in enumerate(dense_matrix):
        print(f"  文本 {doc_idx + 1}: {doc[word_index]:.4f}")
else:
    print(f"詞語 '{target_word}' 不在詞彙表中。")
```
